# SpaceX Falcon 9 Interactive Map of Landings

First we import the dependencies.

```python
import pandas as pd
import folium
```

We need to read the dataset.

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/api/dataset_part_2.csv')

df.head()
```

Now we can display the map using Folium.

First we want the folium map to show the color-labeled launch outcomes on the map.

```python
from folium import plugins

world_map = folium.Map(location=[28, -80], zoom_start=4)
launchSites = plugins.MarkerCluster().add_to(world_map)

for lat, lng, outcome in zip(df.Latitude, df.Longitude, df.Class):
  if outcome == 1:
    markerColor = 'green'
  else:
    markerColor = 'red'
  launchSites.add_child(
    folium.CircleMarker(
      location=[lat, lng],
      popup=str(outcome),
      color=markerColor
    )
  )

world_map.add_child(launchSites)
```
